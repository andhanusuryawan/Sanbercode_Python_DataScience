```
kawin = {
    'Menikah' : 1,
    'Belum Pernah Menikah': 0,
    'Cerai' : 4,          
    'Janda' : 5,                
    'Berpisah' : 3,          
    'Menikah LDR' : 2,
}

kelas = {
    'Wiraswasta' : 8,
    'Pekerja Bebas Bukan Perusahan' : 7,
    'Pemerintah Lokal' : 6,
    '?' : 5,
    'Pemerintah Negara' : 4,
    'Pekerja Bebas Perusahaan' : 3, 
    'Pemerintah Provinsi' : 2, 
    'Tanpa di Bayar' : 1,
    'Tidak Pernah Bekerja' : 0,
}

pendidikan = {
    'SMA' : 8,
   'Pendidikan Tinggi' : 9,
    'Sarjana' : 12,
    'Master' : 13,
    'D4' : 11,
    '11th' : 6,
    'D3' : 10,
    '10th' : 5,
    '7th-8th' : 3,
    'Sekolah Professional' : 15,
    '9th' : 4,
    '12th' : 7,
    'Doktor' : 14,
    '5th-6th' : 1,
    '1st-4th' : 0,
    'SD' : 2,
}

gawe = {
    'Spesialis' : 0,
    'Perbaikan Kerajinan' : 1,
    'Ekesekutif Managerial' : 2,
    'Pemuka Agama' : 3,
    'Sales' : 4,
    'Servis Lainnya' : 5,
    'Mesin Inspeksi' : 6,
    '?' : 7,
    'Supir' : 8,
    'Pembersih' : 9,
    'Petani' : 10,
    'Tech-support' : 11,
    'Penjaga' : 12,
    'Asisten Rumah Tangga' : 13,
    'Tentara' : 14,
}
gaji = {
    '<=7jt' : 0,
    '>7jt' : 1
}
```

```
df_train['Kelas Pekerja'] = df_train['Kelas Pekerja'].replace(kelas)
df_train['Pendidikan'] = df_train['Pendidikan'].replace(pendidikan)
df_train['Status Perkawinan'] = df_train['Status Perkawinan'].replace(kawin)
df_train['Pekerjaan'] = df_train['Pekerjaan'].replace(gawe)
df_test['Kelas Pekerja'] = df_test['Kelas Pekerja'].replace(kelas)
df_test['Pendidikan'] = df_test['Pendidikan'].replace(pendidikan)
df_test['Status Perkawinan'] = df_test['Status Perkawinan'].replace(kawin)
df_test['Pekerjaan'] = df_test['Pekerjaan'].replace(gawe)
df_test=pd.get_dummies(df_test, columns=['Jenis Kelamin'])
df_train=pd.get_dummies(df_train, columns=['Jenis Kelamin'])
df_train['Gaji'] = df_train['Gaji'].replace(gaji)
```

```
df_train
```

```
from sklearn.model_selection import train_test_split
x = df_train.drop('Gaji', axis=1)
y = df_train['Gaji']
x_baru = df_test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
```

```
x_baru
```

```
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
```

```
def train_model(k,w):
    model_knn = KNeighborsClassifier(n_neighbors=k, weights=w)
    model_knn.fit(x_train, y_train)
    model_knn.predict(x_test)
    return model_knn.score(x_test, y_test)

list_uniform = []
list_distance = []
for k in range(1, 50, 1):
  list_uniform.append(train_model(k,'uniform'))
  list_distance.append(train_model(k,'distance'))
```

```
list_uniform
```

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16,8))

k = range(1,50,1)
ax.plot(k, list_uniform, marker='x',color='b', label='Uniform score')
ax.plot(k, list_distance, marker='o',color='y', label='Distance score')
ax.set_xlabel('K-Value')
ax.set_ylabel('Score')
ax.set_title('Accuracy: Uniform scores vs Distance scores')
plt.legend()
plt.show()
```
