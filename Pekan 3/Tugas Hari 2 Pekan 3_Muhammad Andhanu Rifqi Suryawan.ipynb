Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np


x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x, cos, linestyle='-', color='r')
ax.plot(x, cos3, linestyle='--', color='g')
ax.plot(x, sin,  linestyle='-.', color='b')
ax.plot(x, sin2,  marker='x', linestyle= 'none', color='y')
ax.set_xlabel('X Label Axis')
ax.set_ylabel('Y Label Axis')
ax.set_title('Grafik Cos Sin')
plt.show()
```
