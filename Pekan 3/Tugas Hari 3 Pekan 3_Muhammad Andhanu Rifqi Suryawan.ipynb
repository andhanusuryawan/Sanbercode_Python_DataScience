```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True,sharex=True, figsize=(12, 8))
ax[0].plot(x, tan, color='b', label='tan')
ax[1].plot(x, sin, color='r', label='sin')
ax[2].plot(x, cos, color='g', label='cos')
ax[0].set_title('Grafik tan')
ax[1].set_title('Grafik sin')
ax[2].set_title('Grafik cos')
fig.legend()
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig1, axx = plt.subplots(figsize=(16,8))
axx.plot(x, y, color='b')
ins1 = axx.inset_axes([-0.2, -0.2, 0.4, 0.4])
ins2 = axx.inset_axes([0.8, -0.2, 0.4, 0.4])
ins3 = axx.inset_axes([-0.2, 0.8, 0.4, 0.4])
ins4 = axx.inset_axes([0.8, 0.8, 0.4, 0.4])
ins1.plot(x, y2, color='r')
ins1.plot(x, y, color='b')
ins2.plot(x, y3, color='g')
ins2.plot(x, y, color='b')
ins3.plot(x, y5, color='y')
ins3.plot(x, y, color='b')
ins4.plot(x, y4, color='k')
ins4.plot(x, y, color='b')
axx.set_title('Grafik Cosx')
ins1.set_title('Grafik Cosx^2')
ins2.set_title('Grafik Cosx^3')
ins3.set_title('Grafik Cosx^5')
ins4.set_title('Grafik Cosx^4')
```
