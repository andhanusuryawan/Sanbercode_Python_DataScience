<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
1. Pada Series, index pada data bisa kita ubah urutannya dan bisa diubah dengan string, sementara list tidak bisa diubah indexnya. Index dictionary dapat menggunakan kata kunci, bukan dengan urutan angka ataupun huruf.
2. Series adalah objek yang hanya terdiri dari satu dimensi dan mampu menyimpan berbagai jenis file seperti integer, string, dan sebagainya. Sementara dataframne terdiri dari dua dimensi yang terdiri dari index baris dan index kolom.

```python
# code here
import pandas as pd

a = pd.Series(['a', 'b', 'c', 'd'], index = ['satu', 'dua', 'tiga', 'empat'])
print (a)
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
import pandas as pd
rata = pd.DataFrame(data, columns=['price']).mean()
print(rata)
standar = pd.DataFrame(data, columns=['depth']).std()
print(standar)
maks = pd.DataFrame(data, columns=['carat']).max()
print(maks)
```
