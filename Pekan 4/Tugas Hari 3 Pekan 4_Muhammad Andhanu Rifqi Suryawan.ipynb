```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
df['survival_status'].value_counts()
```

```python
# Code here
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
x = df.drop('survival_status', axis=1)
y = df['survival_status']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
```

```python
model_knn=KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(x_train, y_train)
```

```python
y_pred=model_knn.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
model_knn=KNeighborsClassifier(n_neighbors=5)
hasil = cross_val_score(model_knn, x, y,scoring='roc_auc', cv=5)
```

```python
hasil.mean()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_pred, labels=[2,1])
```

```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
# Code here
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, cross_val_score

x = df.drop('survival_status', axis=1)
y = df['survival_status']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
model_knn=KNeighborsClassifier()
```

```python
parapara = {'n_neighbors': np.arange(5, 50), 'weights':['distance', 'uniform']}
gscv = GridSearchCV(model_knn, param_grid=parapara, scoring='roc_auc', cv=5)
gscv.fit(x, y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
gscv.classes_
```

```python
gscv.predict_proba(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
pohon = DecisionTreeClassifier()
```

```python
param={'criterion': ['gini', 'entropy'], 'splitter':['best', 'random'], 'max_depth': np.arange(2, 20)}
rscv=RandomizedSearchCV(pohon, param, scoring='roc_auc', cv=5)
rscv.fit(x,y)
```
